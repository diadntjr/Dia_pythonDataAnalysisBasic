```python
 # read csv file
import csv
f = open('file1.csv', 'r', encoding="utf-8")
rdr = csv.reader(f, delimiter=",", doublequote=True)
for line in rdr:
    print(line)
f.close()
```

```python
 # read csv file
import csv
f = open('file3.csv', 'r', encoding="utf-8")
rdr = csv.reader(f, delimiter=",", doublequote=True)
for row in rdr:
    if '연산제9동' in row[0]:
        for i in row[0:]:
            print(i)
f.close()
```

```python
import matplotlib.pyplot as plt

plt.plot()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('file3.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=",")
result = []
region = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요')

for row in data:
    if region in row[0]:
        for i in row[3:]:
            result.append(int(i))
        
plt.style.use('ggplot')
plt.rc('font', family = 'NanumGothic')

plt.plot(result)
plt.show()
```
